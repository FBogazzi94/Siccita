```python
import pandas as pd
import glob
import os
import datetime
import math
```

#corriere della sera import - cds

```python

list_files_cds = sorted(glob.glob(os.path.join('Corriere_della_sera_*'))) #ordine caricamento files è importante per fill data
df_cds = pd.DataFrame()
columns = ['title','date','author','text','bold_words']

for file in list_files_cds:
  df_cds_tmp = pd.read_csv(file)
  df_cds = pd.concat([df_cds, df_cds_tmp], axis = 0) #concat along rows
df_cds.columns = columns
```

```python
df_cds
```

```python
def test_data_tmp(date):
  date.apply(lambda x : print(x))
```

```python
test_data_tmp(df_cds['date'])
```

```python
### pulizia data ###
#per un caso solo abbiamo il nome dell'autore dentro la data. per ora per semplicità lo escludo a mano

# 1. elimino tutto ciò che viene dopo i seguenti caratteri: ["(", "-", ",", "|"] e creo una lista splittata per spazio
date_tmp = df_cds['date'].apply(lambda x : x.replace("di Ferruccio de Bortoli","").split("(")[0].split(",")[0].split("|")[0].split("-")[0].split(" ") if isinstance(x, str) else None)
test_data_tmp(date_tmp)
```

```python
month_dict = {
    'gen' : 1,
    'feb' : 2,
    'mar' : 3,
    'apr' : 4,
    'mag' : 5,
    'giu' : 6,
    'lug' : 7,
    'ago' : 8,
    'set' : 9,
    'ott' : 10,
    'nov' : 11,
    'dic' : 12
  }
```

```python
# il primo elemento della lista è il giorno, il secondo è il mese e il terzo l'anno. tutto quello che viene dopo si può buttare
day = list(date_tmp.apply(lambda x : int(x[0]) if x != None else None))
month = date_tmp.apply(lambda x : int(month_dict[x[1][0:3].lower()]) if x != None else None)
year = date_tmp.apply(lambda x : int(x[2]) if x != None else None)
date_tuple = list(zip(year, month, day))
date_tuple_old = list(zip(year, month, day))

#fill valori vuoti. prendo il precedente
for i in range(1,len(date_tuple)):
  if all(math.isnan(value) for value in date_tuple[i]):
    date_tuple[i] = date_tuple[i-1]

#converto in data
date_list = [datetime.date(int(anno), int(mese), int(giorno)) for anno, mese, giorno in date_tuple]
```

```python
df_cds['date_cleaned'] = pd.to_datetime(date_list)
df_cds = df_cds[df_cds['date_cleaned'].dt.year >= 2023].drop_duplicates()
df_cds.to_excel('corriere_della_sera_processed.xlsx', index = False)
df_cds = df_cds.drop(['date','author','bold_words'],axis = 1)
df_cds['giornale'] = 'Corriere Della Sera'
df_cds
```

FATTO QUOTIDIANO

```python
dates = []

df_fatto = pd.read_excel('Estrazione_Il_Fatto_Quotidiano.xlsx')
df_fatto['Date']
```

```python
#convert date str into list
dates_list_tmp = []
for date in df_fatto['Date']:
    if isinstance(date,str):
        date_formatted = []
        dates_list_tmp.append(date.strip().replace('|','').split(' '))
    else:
        dates_list_tmp.append(date)
dates_list_tmp
```

```python
month_dict = {
    'gen' : 1,
    'feb' : 2,
    'mar' : 3,
    'apr' : 4,
    'mag' : 5,
    'giu' : 6,
    'lug' : 7,
    'ago' : 8,
    'set' : 9,
    'ott' : 10,
    'nov' : 11,
    'dic' : 12
}

date_tmp = []
#check se ci sono elementi spazio vuoto
for date in dates_list_tmp:
    date_formatted = []
    if isinstance(date, list):
        date_formatted = []
        for date_part in date:
            if date_part != '':
                date_formatted.append(date_part)
        date_tmp.append(date_formatted)
    else:
        date_tmp.append(date)
date_tmp
```

```python
mesi = []
giorni = []
anni = []
for date in date_tmp:
    if isinstance(date, list):
        mesi.append(int(month_dict[date[1][0:3].lower()]))
        anni.append(int(date[2]))
        giorni.append(int(date[0]))
    else:
        mesi.append(math.nan)
        anni.append(math.nan)
        giorni.append(math.nan)

date_tuple = list(zip(giorni, mesi, anni))

#fill valori vuoti. prendo il precedente
for i in range(1,len(date_tuple)):
    if all(math.isnan(value) for value in date_tuple[i]):
        date_tuple[i] = date_tuple[i-1]

df_fatto['date_cleaned'] = [datetime.date(int(anno), int(mese), int(giorno)) for giorno, mese, anno in date_tuple]
```

```python
df_fatto['date_cleaned'] = pd.to_datetime(df_fatto['date_cleaned'])
df_fatto = df_fatto[df_fatto['date_cleaned'].dt.year >= 2023].drop_duplicates()
df_fatto
```

```python
df_fatto.to_excel('fatto_quotidiano_processed.xlsx', index = False)
df_fatto = df_fatto.drop(['URL','Date','Bold Words'],axis = 1)
df_fatto.columns = ['title', 'text', 'date_cleaned']
df_fatto['giornale'] = 'Fatto Quotidiano'
df_fatto
```

```python
#convert date str into list
date_tmp = []
for date in df_repubblica['Date']:
    if isinstance(date,str):
        date_formatted = []
        date_tmp.append(date.strip().replace('|','').split(' '))
    else:
        date_tmp.append(date)
date_tmp
```

```python
mesi = []
giorni = []
anni = []
for date in date_tmp:
    if isinstance(date, list):
        mesi.append(int(month_dict[date[1][0:3].lower()]))
        anni.append(int(date[2]))
        giorni.append(int(date[0]))
    else:
        mesi.append(math.nan)
        anni.append(math.nan)
        giorni.append(math.nan)

date_tuple = list(zip(giorni, mesi, anni))

#fill valori vuoti. prendo il precedente
for i in range(1,len(date_tuple)):
    if all(math.isnan(value) for value in date_tuple[i]):
        date_tuple[i] = date_tuple[i-1]

df_repubblica['date_cleaned'] =  pd.to_datetime([datetime.date(int(anno), int(mese), int(giorno)) for giorno, mese, anno in date_tuple])
```

IL SOLE 24 ORE

```python
dates = []

df_sole_24_ore = pd.read_excel('Estrazione_Sole24ore.xlsx')
df_sole_24_ore
```

```python
#convert date str into list
date_tmp = []
for date in df_sole_24_ore['Date']:
    if isinstance(date,str):
        date_formatted = []
        date_tmp.append(date[:date.find('T')].split('-'))
    else:
        date_tmp.append(date)
date_tmp
```

```python
mesi = []
giorni = []
anni = []
for date in date_tmp:
    if isinstance(date, list):
        mesi.append(int(date[1]))
        anni.append(int(date[0]))
        giorni.append(int(date[2]))
    else:
        mesi.append(1)
        anni.append(1900)
        giorni.append(1)

date_tuple = list(zip(giorni, mesi, anni))


df_sole_24_ore['date_cleaned'] = [datetime.date(int(anno), int(mese), int(giorno)) for giorno, mese, anno in date_tuple]
df_sole_24_ore
```

```python
df_sole_24_ore.to_excel('sole_24_ore_processed.xlsx', index = False)
df_sole_24_ore = df_sole_24_ore.drop(['URL','Date'],axis = 1)
df_sole_24_ore.columns = ['title', 'text', 'date_cleaned']

df_sole_24_ore['giornale'] = 'Il Sole 24 Ore'
df_sole_24_ore
```
